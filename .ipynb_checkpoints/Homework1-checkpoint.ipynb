### Answer 1.3

## Question 3.1

The benefit with tidy data is that it is easy to work with. For this task use the tidy weather dataset from above and choose three months and by **a density and histogram plot** visualize the differences in temperature between 2009 and 2010 for the chosen months in a readable manner. Make sure to print the visuals below. 

```python
# Insert code here for histogram
subdata = tidy_data.query("(month == 'Jun' or month == 'Jul' or month == 'Aug') and (variable == 'Temp2009' or variable =='Temp2010')")
subdata = subdata.assign(year= lambda df: df['variable'].map(lambda var: int(2009) if var == "Temp2009" else int(2010)))

p2 = (ggplot(subdata, aes("value", fill="variable"))
    + geom_histogram(bins=50, alpha=0.5, position = "identity")
    + scale_fill_manual(values=cbPalette, name = "Year", labels=[2009, 2010])
    + labs(x = "Summer temperature at 12:00 [C]", title = "2018 - What a time to be alive")
    + theme_minimal())

p2
# Remember to print the visual 
```

```python
# Insert code here for density plot
p1 = (ggplot(subdata, aes("value", fill="variable"))
    + geom_density(bw=0.5, alpha=0.6)
    + scale_fill_manual(values=cbPalette, name = "Year", labels=[2009, 2010])
    + labs(x = "Temperature", title = "Temperature Difference between 2009 and 2010")
    + theme_minimal())
p1
# Remember to print the visual 
```

Provide a brief motivation for which plot (density and histogram) you find most readable.



_Your answer here_

### Question 3.2

In Lecture2 I used boxplots, error-bars, and violin plots to visualize fluctuations in winter temperature in Västerås. Now using the weather data above plot the temperature per month (like in Fig. 9.8 [here](https://clauswilke.com/dataviz/boxplots-violins.html)) using error-bars, boxplots and violin plots with data points. For each month plot the temperature for 2009 and 2010 next to each other (see example on webpage). 

```python
# Insert code here using standard error
eb = pd.DataFrame({"month": [],
                   "year": [],
                   "mean": [],
                   "sd": []})
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
year = ['Temp2009', 'Temp2010']
for i in month:
    for j in year:
        x = tidy_data.query("(month == @i) and (variable == @j)")
        #df = pd.concat([df, pd.DataFrame({"month": i, "year": j, "mean": np.mean(x.value), "sd": np.std(x.value)})])
        eb.loc[len(eb.index)] = [i, j, np.mean(x.value), np.std(x.value)] 

p = (ggplot(eb, aes(fct_inorder(eb["month"]), "mean"))
    + geom_point(aes(color = "year"), position = position_dodge(0.3)) 
    + geom_errorbar(aes(ymin = "mean - sd", ymax = "mean + sd", color="year"), position=position_dodge(0.3), width=0.2)
    + labs(y = "Temperature", x = "month", title="Temperature for 2009 and 2010")
    + scale_fill_manual(values = cbPalette, name="Year") 
    + theme_minimal())

p
# Remember to print the visual 
```

```python
# Insert code here using boxplots
box = tidy_data.query("(variable == 'Temp2009' or variable == 'Temp2010')")

p1 = (ggplot(box, aes(fct_inorder(box["month"]), y= "value")) 
     + geom_boxplot(aes(fill = "variable"), position = position_dodge(0.9) )
     + labs(y = "Temperature", x = "month", title="Temperature for 2009 and 2010")
     + scale_fill_manual(values = cbPalette, name="Year")
     + theme_minimal())

p1
# Remember to print the visual 
```

```python
# Insert code here using violin plot with dots
p1 = (ggplot(box, aes(fct_inorder(box["month"]), y= "value"))
     + geom_violin(aes(fill = "variable"), trim = True, position = position_dodge(0.9) ) 
     + labs(y = "Temperature", x = "month", title="Temperature for 2009 and 2010")
     + scale_fill_manual(values = cbPalette, name="Year")
     + theme_minimal()     )
p1
# Remember to print the visual 
```

Briefly discuss the drawback with each approach (standard errors, boxplots, and violin plots). 

__Write answer here__

For standard errors, it is not obvious what the points and error bars represent. This may confuse the readers.

For boxplots, it is hard to observe the detailed shape of distribution.

For violin plots, the quantitative value for the mean is not provided. 


### Question 3.3

Some people argue that visuals should be as minimal as possible. On the webpage you have a bad example of a minimalist way to plot the min, median, and max temperature of each month in 2009 and 2010 using the dataset in Question 3.2 (where you hopefully have filtered out daily temperatures for 2009 and 2010). By **only** using geom\_point recreate the visual (you do not need to recreate the title). **You are not allowed to transform the dataset**.

*Hint - Remember the stat argument in ggplot*

```python
# Insert code here 

p1 = (ggplot(box, aes(fct_inorder(box["month"]) , 'value'))
        + geom_point(position = "jitter")
        + geom_point(stat = "summary",
            fun_y = np.mean,
            position = "identity",
            size = 3.0,
            color="red"))

p1

#
# Remember to print the visual 
```

### Question 4.1 

Barplots are the workhorse for plotting amounts. Create a dataframe with the values from Tab. 6.1 in the [course book](https://clauswilke.com/dataviz/visualizing-amounts.html), and recreate Fig. 6.1 and Fig. 6.3 in the course book.

```python
# Insert code here for Fig. 6.1
movies = pd.DataFrame({"Rank": [1, 2, 3, 4, 5],
                      "Title": ["Star Wars: The Last Jedi", "Jumanji: Welcome to the Jungle", "Pitch Perfect 3", "The Greatest Showman", "Ferdinand"],
                      "Weekend_gross": [71565498, 36169328, 19928525, 8805843, 7316746]})


pos = movies["Title"].values[np.argsort(movies["Weekend_gross"])]
p = (ggplot(movies, aes("Title", "Weekend_gross")) 
  + geom_bar(stat="identity", width = 0.5, position = position_dodge(0.7))
  + geom_text(aes(label = "Weekend_gross"))
  + labs(x = "Movies", y = "Weekend gross")
  + scale_x_discrete(limits=pos[::-1],expand = (0, 0.4))
  + theme_minimal()
  + theme(axis_text_x = element_text(angle = 45, hjust=0.1), 
          legend_position="none"))
p
# Remember to print the visual 
```

```python
# Insert code here for Fig. 6.2
p = (ggplot(movies, aes("Title", "Weekend_gross")) 
  + geom_bar(stat="identity", width = 0.5, position = position_dodge(0.7))
  + geom_text(aes(label = "Weekend_gross"))
  + labs(x = "Movies", y = "Weekend gross")
  + scale_x_discrete(limits=pos[::-1],expand = (0, 0.4))
  + coord_flip()
  + theme_minimal()
  + theme(axis_text_x = element_text(angle = 45, hjust=0.1), 
          legend_position="none"))
p
# Remember to print the visual 
```

### Question 4.2 

Often we want to highlight a specific column in a barplot. Using the same layout as in the most recent plot highlight the column of Jumanji in orange, and keep the remaining columns grey.

```python
# Insert code here 
movies = movies.assign(
            highlight = lambda dataframe: dataframe['Title'].map(lambda title: True if title == "Jumanji: Welcome to the Jungle" else False) 
            )
p = (ggplot(movies, aes("Title", "Weekend_gross", fill= "highlight")) 
  + geom_bar(stat="identity", width = 0.5, position = position_dodge(0.7))
  + geom_text(aes(label = "Weekend_gross"))
  + labs(x = "Movies", y = "Weekend gross")
  + scale_x_discrete(limits=pos[::-1],expand = (0, 0.4))
  + coord_flip()
  + scale_fill_manual( values = ["grey", "orange"], guide = False )
  + theme_minimal()
  + theme(axis_text_x = element_text(angle = 45, hjust=0.1), 
          legend_position="none"))
p

# Remember to print the visual 
```
