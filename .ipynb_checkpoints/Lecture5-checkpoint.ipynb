## Uncertainty 

Let us start with Spanish election data. Here like a normal person adding an error-bar makes all the difference. 

```python
data_catalonia = pd.DataFrame({"percent" : [45.3, 44.4, 10.2], 
                               "low_CI" : [45.3 - 2.95, 44.4 - 2.95, 10.2 - 2.95], 
                               "upp_CI" : [45.3 + 2.95, 44.4 + 2.95, 10.2 + 2.95],
                               "answer" : pd.Categorical(["No", "Yes", "No answer"], 
                                                          categories=["No", "Yes", "No answer"])})

p1 = (ggplot(data_catalonia, aes("answer"))
    + geom_linerange(aes(ymin = 0.0, ymax = "percent"), size=2.0, position=position_dodge(0.2))
    + geom_crossbar(aes(y = "percent", ymin = "percent", ymax = "percent"), width=0.2, size=1.0)
    + geom_text(aes(label="percent", y="percent"), 
                stat="identity", 
                nudge_y=3.0)
    + scale_y_continuous(expand=(0.0, 2.0))
    + labs(x = "", y = "Percentage", title = "Do you want Catalonia independence?")
    + theme_minimal()
    + theme(panel_grid_major_x=element_blank(),
            axis_text_x = element_text(color="black", size=12)))     

p2 = (ggplot(data_catalonia, aes("answer"))
    + geom_errorbar(aes(ymin="low_CI", ymax="upp_CI"), width=0.1, size=1.0)
    + geom_text(aes(label="percent", y="percent"), 
                stat="identity", 
                nudge_y=4.5)
    + scale_y_continuous(expand=(0.0, 2.0))
    + labs(x = "", y = "Percentage", title = "Do you want Catalonia independence?")
    + geom_crossbar(aes(y = "percent", ymin = "percent", ymax = "percent"), width=0.2, size=0.7)
    + theme_minimal()
    + theme(panel_grid_major_x=element_blank(),
            axis_text_x = element_text(color="black", size=12)))                                                 

p1.save("Figs/Poll_bad.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p2.save("Figs/Poll_good.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)

[p1, p2]
```

Always mark what the error-bars correspond to. However, even if this is done many still have a hard to interpret them. Here I think it can help to highlight that there are different levels of uncertainty. The fuzzy plot below is a tricky call, but I think it is a good approach when communicating towards a lay-audience.

```python
p = (ggplot(chocolate_data_sum, aes(x = "Company_location")) 
  + geom_errorbar(aes(ymin = "CI_low", ymax = "CI_high", color = "CI_percent"), width=0.15)
  + geom_errorbar(aes(ymin = "CI_low", ymax = "CI_high", size = "CI_percent", color="CI_percent", width=0.0))
  + scale_color_manual(values = ["#a6bddb", "#3690c0", "#045a8d"], name = "Confidence level")
  + scale_size_manual(values = [1.5, 3.0, 4.5], name = "Confidence level")
  + labs(x = "", y = "Rating", title = "Rating of chocolate bars from different countries")
  + theme_minimal()
  + theme(panel_grid_major_x=element_blank(),
          axis_text_x = element_text(size=12)))

p.save("Figs/Chocolate_CI_levels.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p
```

```python
# If we want to do a "fuzzy" plot we create "lines" and for each line associated 
data_fuzzy_tmp = chocolate_data_sum.query("CI_percent == '99%'")

line_color = np.concatenate([np.linspace(0, 50, 50), np.linspace(50, 0, 50)])
data_fuzzy = pd.DataFrame()

for i in range(data_fuzzy_tmp.shape[0]):
    val_min, val_max = data_fuzzy_tmp["CI_low"].values[i], data_fuzzy_tmp["CI_high"].values[i]
    df_tmp = pd.DataFrame({"val_line" : np.linspace(val_min, val_max, 100), 
                        "color_line" : line_color}) 
    df_tmp["Company_location"] = data_fuzzy_tmp["Company_location"].values[i]
    data_fuzzy = pd.concat([data_fuzzy, df_tmp])


p = (ggplot(data_fuzzy, aes("Company_location", "val_line"))
    + geom_line(aes(color="color_line"), size=11.0, alpha=0.7)
    + geom_crossbar(data=data_sum_95, mapping=aes(y = "mean", ymin = "mean", ymax = "mean"), width=0.26, size=1.0)
    + scale_color_cmap("Blues")
    + labs(x = "", y = "Rating", title = "Rating of chocolate bars from different countries")
    + theme_minimal()
    + theme(legend_position="none")
    + theme(panel_grid_major_x=element_blank(),
          axis_text_x = element_text(size=12)))

p.save("Figs/Fuzzy.png", dpi=300, width=BASE_WIDTH, height=BASE_HEIGHT)
p
```

It is quite common to fit trend-lines to data. When doing this you should always highlight the uncertainty in said curve fit. Now let say we want to fit the data below to the function $N_{mod} = N_0 \mathrm{exp}(rt)$ by estimating $N_0$ and $r$ via a maximum likelihood estimate assuming normal measurement noise; $N_{obs} \sim \mathcal{N}(N_{mod}, \sigma^2)$

```python
data_obs = pd.read_csv("Data/Data_obs.csv")
data_fit = pd.read_csv("Data/Sum_curve_fit.csv")
data_traces = pd.read_csv("Data/Traces_fit.csv")

data_fit = data_fit.assign(Confidence_level = pd.Categorical(data_fit["Confidence_level"], 
                                                       categories=["99%", "90%", "80%"]))

data_fit_use = data_fit.query("t <= 5.1")
data_fit_use_95 = data_fit_use.query("Confidence_level == '90%'")
data_traces_use = data_traces.query("t <= 5.1")


p1 = (ggplot(data_obs, aes("t", "N"))
    + geom_point()
    + labs(x = "Time [h]", y = "Population size", title = "Observed data on population growth")
    + theme_bw())

p2 = (ggplot(data_fit_use, aes("t", "N"))
    + geom_point(data=data_obs, mapping=aes("t", "N"))
    + geom_line(aes("t", "median"))
    + labs(x = "Time [h]", y = "Population size", title = "Observed data on population growth")
    + theme_bw())

p3 = (ggplot(data_fit_use_95, aes("t"))
    + geom_ribbon(aes(ymin = "Lower_CI", ymax="Upper_CI"), alpha=0.6, fill="#74a9cf")
    + geom_point(data=data_obs, mapping=aes("t", "N"))
    + geom_line(aes("t", "median"))
    + labs(x = "Time [h]", y = "Population size", title = "Fitted model to population growth data")
    + theme_bw())

col_use = ["#0570b0", "#74a9cf", "#d0d1e6"]
p4 = (ggplot(data_fit_use, aes("t"))
    + geom_ribbon(aes(ymin = "Lower_CI", ymax="Upper_CI", fill="Confidence_level"), alpha=0.6)
    + geom_point(data=data_obs, mapping=aes("t", "N"))
    + geom_line(aes("t", "median"))
    + labs(x = "Time [h]", y = "Population size", title = "Fitted model to population growth data")
    + scale_fill_manual(values = col_use[::-1])
    + theme_bw())

p5 = (ggplot(data_fit_use_95, aes("t"))
    + geom_ribbon(aes(ymin = "Lower_CI", ymax="Upper_CI"), alpha=0.4)
    + geom_line(data=data_traces_use, mapping=aes("t", "N_pred", group="index"), size=0.3, color=col_use[0])
    + geom_point(data=data_obs, mapping=aes("t", "N"))
    + labs(x = "Time [h]", y = "Population size", title = "Fitted model to population growth data")
    + theme_bw())


p1.save("Figs/Just_data.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p2.save("Figs/Data_line.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p3.save("Figs/Data_bands.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p4.save("Figs/Data_several_bands.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p5.save("Figs/Data_band_line.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)

[p1, p2, p3, p4, p5]
```

Sometimes our outcome is not discrete, rather it might be a continuous density (e.g. often when doing Bayesian inference). Instead of saying there is a 30% chance that the temperature goes below 0-degree showing the density can be more efficient to simply plot the density. Adding dots can further help, I have done this in a R-script (as this is quite the complicated operation to do with ggplot).

```python
# Generate density data 
norm_dist = sp.stats.norm(loc=1.2, scale=0.9)
T_val = np.linspace(-2.5, 5.0, 300)
pdf_val = norm_dist.pdf(T_val)


data_pdf = pd.DataFrame({"Temp" : T_val,
                         "pdf" : pdf_val})
data_pdf["below0"] = "No"
data_pdf["below0"][data_pdf["Temp"].to_numpy() < 0.0] = "Yes"

col_use = ["#d1e5f0", "#fddbc7"]
p = (ggplot(data_pdf, aes("Temp", "pdf", fill="below0"))
    + geom_area(color="black")
    + geom_vline(xintercept=0.0, size=1.0)
    + scale_y_continuous(expand=(0, 0))
    + scale_fill_manual(values=col_use[::-1], name = "Below 0")
    + labs(x = "Temperature [C]", y = "Density", title = "13% chance temperature is below 0")
    + theme_minimal()
    + theme(legend_position="none"))

p.save("Figs/Dens_prob.png", width=BASE_WIDTH*1.5, height=BASE_HEIGHT, dpi=300)          
p 
```

```python
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
data_forest = pd.read_csv("Data/Forest_tidy.csv")

# EPSG:6933
world_lambert = world.to_crs("EPSG:6933")
data_world_forest = world_lambert.merge(data_forest, left_on="name", right_on="Country")
p1 = (ggplot(data_world_forest, aes(fill="2019_forest"))
    + geom_map()
    + labs(title="Forest coverage : EPSG:6933")
    + scale_fill_cmap("Greens", name = "Forest coverage [%]") 
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

# Mercator
world_mercator = world.to_crs("EPSG:3395")
data_world_forest = world_mercator.merge(data_forest, left_on="name", right_on="Country")
p2 = (ggplot(data_world_forest, aes(fill="2019_forest"))
    + geom_map()
    + labs(title="Forest coverage : Mercator")
    + scale_fill_cmap("Greens", name = "Forest coverage [%]")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

# Mercator
world_wsg84 = world.to_crs("EPSG:4326")
data_world_forest = world_wsg84.merge(data_forest, left_on="name", right_on="Country")
p3 = (ggplot(data_world_forest, aes(fill="2019_forest"))
    + geom_map()
    + labs(title="Forest coverage : WGS84")
    + scale_fill_cmap("Greens", name = "Forest coverage [%]")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

p1.save("Figs/Maps_forest_lambert.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p2.save("Figs/Maps_forest_mercator.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p3.save("Figs/Maps_forest_WSG84.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)

[p1, p2, p3]
```

```python
map_USA = gpd.read_file("Data/cb_2018_us_state_500k.shp")

starbucks_stores = pd.read_csv("Data/Startbucks_location.csv")
us_stores = starbucks_stores[starbucks_stores.Country=="US"]

# Remove non-mainland terratories (+ Alaska)
us_stores = us_stores.loc[us_stores["State/Province"] != "VI"]
us_stores = us_stores.loc[us_stores["State/Province"] != "MP"]
us_stores = us_stores.loc[us_stores["State/Province"] != "GU"]
us_stores = us_stores.loc[us_stores["State/Province"] != "HI"]
us_stores = us_stores.loc[us_stores["State/Province"] != "AK"]

# Prepare dataframe
us_stores_statewise_tmp = us_stores.groupby("State/Province").count()[["Store Name"]].rename(columns={"Store Name":"Count"})
us_stores_statewise_tmp = us_stores_statewise_tmp.reset_index()
us_stores_statewise =  map_USA.merge(us_stores_statewise_tmp, left_on="STUSPS", right_on="State/Province")

us_stores_statewise["Count_norm"] = us_stores_statewise["Count"] / us_stores_statewise["geometry"].area 

p1 = (ggplot(us_stores_statewise)
    + geom_map(aes(fill="Count"))
    + labs(title = "Number of Starbucks per sqm2 in USA -(Alaska and Hawaii)")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

p2 = (ggplot(us_stores_statewise)
    + geom_map(fill=col_highlight[0])
    + geom_point(data=us_stores, mapping=aes(x="Longitude", y="Latitude"), color=col_highlight[1], alpha=0.3)
    + labs(title = "Number of Starbucks in USA -(Alaska and Hawaii)")
    + labs(x = "", y = "")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

us_stores_statewise["x"] = us_stores_statewise["geometry"].centroid.map(lambda p: p.x)
us_stores_statewise["y"] = us_stores_statewise["geometry"].centroid.map(lambda p: p.y)
p3 = (ggplot(us_stores_statewise)
    + geom_map(fill=col_highlight[0])
    + geom_point(aes(x="x", y="y", size="Count"), color=col_highlight[1])
    + labs(title = "Number of Starbucks in USA -(Alaska and Hawaii)")
    + labs(x = "", y = "")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))


p1.save("Figs/Starbucks1.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p2.save("Figs/Starbuck2.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)

# Homework - Asian market
[p1, p2, p3]
```

```python
# Let us play around a bit with different color-schemes 



quantile_val = np.quantile(us_stores_statewise["Count"], [0.2, 0.4, 0.6, 0.8, 0.95])

conditions = [us_stores_statewise["Count"].lt(quantile_val[0]), 
              (us_stores_statewise["Count"].ge(quantile_val[0]) & us_stores_statewise["Count"].lt(quantile_val[1])), 
              (us_stores_statewise["Count"].ge(quantile_val[1]) & us_stores_statewise["Count"].lt(quantile_val[2])), 
              (us_stores_statewise["Count"].ge(quantile_val[2]) & us_stores_statewise["Count"].lt(quantile_val[3])), 
              (us_stores_statewise["Count"].ge(quantile_val[3]) & us_stores_statewise["Count"].lt(quantile_val[4])), 
              us_stores_statewise["Count"].ge(quantile_val[4])]
choices = ["Less than 34", "35 to 92", "93 to 215", "216 to 366", "367 to 731", "More than 731"]
us_stores_statewise["N_stores_quant"] = pd.Categorical(np.select(conditions, choices), 
                                                       categories=choices[::-1])

col_use = ["#dadaeb", "#bcbddc", "#9e9ac8", "#807dba", "#6a51a3", "#4a1486"]
p = (ggplot(us_stores_statewise)
    + geom_map(aes(fill="N_stores_quant"))
    + labs(title = "Number of Starbucks per state in USA -(Alaska and Hawaii)")
    + scale_fill_manual(values=col_use[::-1], name = "Number of stores")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

p.save("Figs/Starbuck_bin.png", width=BASE_WIDTH, height=BASE_HEIGHT, dpi=300)
p
```
