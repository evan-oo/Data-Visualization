### Question 2.1

Coffee is undoubtedly a popular beverage (albeit inferior to tea :). On Canvas I have uploaded a dataset with ratings for coffee beans from six different countries. 

Here we are interested in identifying which country on average has the best quality beans. Create two plots, where in i) you plot the mean ($\mu$) and standard deviation for each country (using errorbars), and ii) where you plot the mean and standard error of the mean estimate. Briefly discuss the drawbacks with error-bars. 

As a reminder. For country $j$ with coffee ratings $x^{(j)}_1, \ldots, x^{(j)}_{n_j}$ the sample mean is given by 

$$
\hat{\mu}^{(j)} = \sum_{i = 1}^{n_j} x_i^{(j)},
$$

sample standard deviation by

$$
\hat{\sigma}^{(j)} = \sqrt{\frac{1}{n - 1} \sum_{i = 1}^{n_j} (x_i^{(j)} - \hat{\mu}^{(j)})^2},
$$

and sample standard error for $\hat{\mu}^{(j)}$

$$
\hat{\sigma}_{\hat{\mu}^{(j)}} = \frac{\hat{\sigma}^{(j)}}{\sqrt{n}}
$$

and a confidence interval with confidence level $1 - \alpha$ is given by;

$$
\hat{\mu}^{(j)} \pm t_{n-1}(1 - \alpha / 2) \hat{\sigma}_{\hat{\mu}^{(j)}}
$$

where $t_{n-1}$ is the t-distribution with $n-1$ degree of freedom. For confidence level $1 - \alpha=0.95$ and $n = 30$ we have that $t_{29}(1 - 0.025) \approx 2.05$.

```python
# Insert code here for visual 1
coffee = pd.read_csv("Data/HW3/Coffee_tidy.csv")
countries = ["Brazil", "Colombia", "Guatemala", "Kenya", "Mexico", "Taiwan"]
coffee_count = []
coffee_mean = []
coffee_std = []
for c in countries:
    c = coffee.query("(country_of_origin==@c)")
    coffee_count.append(len(c))
    coffee_mean.append(np.mean(c.total_cup_points))
    coffee_std.append(np.std(c.total_cup_points))

df = pd.DataFrame({"country": countries,
                   "mean":coffee_mean,
                   "sd": coffee_std,
                   "count": coffee_count,
                   "se": coffee_std/np.sqrt(coffee_count)})
display(df)

p2 = (ggplot(df, aes("country", "mean"))
  + geom_point(size=3.0)
  + geom_errorbar(aes(ymin="mean - sd", ymax="mean + sd"), width=0.2, size=1.0)
  + labs(x = "", y = "Rating", title = "Rating of coffee from different countries")
  + theme_minimal()
  + theme(panel_grid_major_x=element_blank(),
          axis_text_x = element_text(size=12)))

p2
# Remember to print the visual 
```

```python
# Insert code here for visual 2
p2 = (ggplot(df, aes("country", "mean"))
  + geom_point(size=3.0)
  + geom_errorbar(aes(ymin="mean - se", ymax="mean + se"), width=0.2, size=1.0)
  + labs(x = "", y = "Rating", title = "Rating of coffee from different countries")
  + theme_minimal()
  + theme(panel_grid_major_x=element_blank(),
          axis_text_x = element_text(size=12)))
    
p2
# Remember to print the visual 
```

```python
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# data_plot = asia.merge(starbucks_stores, left_on="iso_a3", right_on="iso_a3") will give you a dataset you can work with

(ggplot(world)
    + geom_map())
```

```python
# Insert code here 
forest1990 = pd.read_csv("Data/HW3/Forest_1990.csv")
forest2019 = pd.read_csv("Data/HW3/Forest_2019.csv")
forest_data = pd.DataFrame({"Country": forest1990.Country, "changed": forest2019.forest - forest1990.forest})


# Mercator
world_mercator = world.to_crs("EPSG:3395")
data_world_forest = world_mercator.merge(forest_data, left_on="name", right_on="Country")
p2 = (ggplot(data_world_forest, aes(fill="changed"))
    + geom_map()
    + labs(title="Forest coverage changed between 1990 and 2019")
    + scale_fill_cmap("BrBG", trans="log10", name = "Forest coverage changed[%]")
    #+ scale_fill_cmap("Greens", name = "Forest coverage changed[%]")
    + theme(panel_background=element_rect(fill = "white"), 
            axis_ticks=element_blank(), 
            axis_text=element_blank()))

p2
# Remember to print the visual 
```
